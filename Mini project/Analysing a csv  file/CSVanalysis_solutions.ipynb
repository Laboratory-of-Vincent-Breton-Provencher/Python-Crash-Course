## 1. Importing and playing with the CSV file

To import the csv file, I suggest using the function `read_csv` from pandas. 

```python
# In this cell, load and display the Dataframe

directory = 'ProbCond_924_2025-05-15_13-37_trialMTX.csv'

dataframe = pd.read_csv(directory)
dataframe
```

```python
# here, extract the column tone_id using both methods

tone_id1 = dataframe['tone_id']
tone_id2 = dataframe.iloc[:,1]

print(tone_id1)
print(tone_id2)
```

The method iloc can also be used to extract the information from a row.

## 2. Computing the average of a row under a given condition

Now that we are experts are manipulating rows and columns, we will next compute the average of one column when another is equal to a given value. To do this, we must

1. Extract the rows where the `tone_id` is equal to 2.
2. Get the associated values for `bl_lickrate`.
3. Compute the average of these rows

```python
equal_to_2 = tone_id1 == 2
bl_lickrates = dataframe['bl_lickrate'][equal_to_2]
average_value = np.mean(bl_lickrates)
```

## 3. Repeat it for many files

Now, say we have many files we wish to compute this for, write a code that will return a list for the averages for every files. 

```python
directories = ['file1', 'file2']
averages = []

for i in range(len(directories)):

    # 1. Load the file
    
    directory = directories[i]
    directory = 'ProbCond_924_2025-05-15_13-37_trialMTX.csv'
    dataframe = pd.read_csv(directory)

    # 2. Extract the relevant columns 

    tone_id = dataframe.iloc[:,1]
    bl_lickrate = dataframe.iloc[:,7]

    # 3. Find columns where tone_id is equal to 2 and the associated values for bl_lickrate

    equal_to_2 = tone_id==2
    bl_lrs = bl_lickrate[equal_to_2]
    ave = np.mean(bl_lrs)

    # 4. store the values in a list

    averages.append(ave)
```

Finaly, write a csv file where the columns are `filename` and `average bl_lickrate`. To do so, first create such a dictionary and then store it as a csv file. 

```python
dictionary = {'filename': directories, 'average bl_lickrate' : averages}
to_save = pd.DataFrame(dictionary)

to_save
```

```python
to_save.to_csv('results.csv')
```

## 4. Displaying the data

 Lastly, we will create a code to display the data. Say we want to display the average by sessions. If our directories list was organized in chronological order, the averages will also be in this order. We can therefore simply plot the avrages we computed earlier. 

```python
plt.figure(figsize=(5,5), layout='constrained')
plt.plot(averages)
plt.xlabel('Sessions')
plt.ylabel('Average bl lickrate')
plt.title('Simplest plot')
plt.savefig('simple_graph.png')
plt.show()
```
